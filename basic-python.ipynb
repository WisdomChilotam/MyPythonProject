```python

def palindrome_num(number):
    str_num = str(number)
    if str_num == str_num[::-1]:
        print(f"Number {str_num}, is a palindrome number")
    else:
        print(f"Number {str_num}, is not a palindrome number")

palindrome_num(1234)
palindrome_num(1221)

```

```python
def exponent(base, exp):
    num = exp
    result = 1
    
    while num > 0:
        result = result * base
        num = num - 1
    print(base, "raises to the power of", exp, 'is:', result)

exponent(2, 5)
exponent
```
