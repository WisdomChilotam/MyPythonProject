### Define a property that must have the same value for every class instance (object)

```python
class Vehicle:

    def __init__(self, name, max_speed, mileage, color):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage
        self.color = color


class Bus(Vehicle):
    pass

class Car(Vehicle):
    pass

School_bus = Bus("School Volvo", 180, 12, "White")
print(School_bus.color, School_bus.name, "Speed:", School_bus.max_speed, "Mileage:", School_bus.mileage)

car = Car("Audi Q5", 240, 18, "Yellow")
print(car.color, car.name, "Speed:", car.max_speed, "Mileage:", car.mileage)
```

### Class inheritance

```python
class Vehicle:
    def __init__ (self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100

class Bus(Vehicle):
    def fare(self):
        amount = super().fare()
        amount += amount * 10/100
        return amount

school_bus = Bus("School Volvo", 12, 50)
print("Total Bus fare is:", school_bus.fare())
```

### Check object is a subclass of a particular class

```python
class Animal:
    pass

class Dog(Animal):
    pass

class Puppy(Dog):
    pass

class Cat:
    pass

print(issubclass(Puppy, Animal))
print(issubclass(Cat, Animal))
print(super(Animal, Dog))
```

### Calculate the area of different shapes using OOP

```python
class Shape:
    def area(self):
        raise NotImplementedError("Area method must be implemented by subclasses")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius**2

class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side * self.side

# Example of polymorphism
shapes = [Circle(5), Square(7), Circle(3), Square(15)]

for shape in shapes:
    print(shape.area())
```
